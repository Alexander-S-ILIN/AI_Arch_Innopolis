## Предобработка данных 

```python
#приводим к змеиному регистру наименования столбцов датафрейма
data.columns = data.columns.str.lower()
```

Заполним пропущенные значения в year_of_release соответствующим найденным значением года для аналогичной игры

```python
#сначала заполним пропуски нулями 
data.year_of_release = data.year_of_release.fillna(value=0) 
```

```python
#приведем столбец year_of_release к целочисленному типу
data.year_of_release = data.year_of_release.astype(int)
```

```python
# выберем в список уникальные наименования, в которых пропуски
data_year_of_release_null = data[data.year_of_release == 0].name.to_list() 
```

```python
# создадим список годов релизов к уникальным наименованиям игр и там где год найден, подставим вместо нулевых значений, найденный по наименованию год релиза.
for name in data_year_of_release_null:
    year = data[(data.name == name) & (data.year_of_release != 0)].year_of_release.to_list()
    if len(year) != 0: 
        data.loc[(data.name == name) & (data.year_of_release == 0), 'year_of_release'] = year[0]
```

```python
data[data.year_of_release==0].count()
```

Часть данных год релиза была восстановлена по наименованию игры. Оставшиеся пропуски целесообазно удалить, чтобы не было нелогичного установления года выпуска. Пропуск в данном столбце мог возникнуть из-за того, что при заполнении данных об игре год релиза забыли указать или точная дата выхода игры точно не определена.

```python
#сохраним значения с неизвестным годом релиза в отдельный датафрейм
data_year_of_release_null = data[data.year_of_release == 0]
#исключим строки из датафрейма, по которым не удалось установить год релиза (составляют менее 1% от данных)
data = data[data.year_of_release != 0]
```

```python
data.info()
```

Удалим строки, по которым не указаны наименование (name) и жанр (genre): по данной ячейке мы не можем рассчитать среднее значение или медиану. Оставить его пустым будет неверным, т.к. непонятно о какой видеоигре идет речь, поэтому мы принимаем решение удалить эти 2 строки, иначе они повлияют на нашу статистику. Жанры не указаны по тем же самым строкам, где не указаны наименования игр.

В столбце user_score выделим значение tbd. Выше мы воспользовались методом info() и обратили внимание, что формат ячейки object. Судя по общему содержанию значений столбца, а так же по описанию столбца (оценка пользователей (максимум 10)) тип данных должен быть float. Поэтому целесообразно считать данное значение (tbd) пустым и изменить на NaN.

```python
# Функция для замены значения 'tbd' на NaN
def replace_wrong_user_scores(wrong_user_scores, correct_user_score):
    data.user_score = data.user_score.replace(wrong_user_scores, correct_user_score)

replace_wrong_user_scores('tbd', np.NaN)
#проверяем уникальные значения в столбце после применения функции
data['user_score'].unique()
```

Значение в столбце user_score скорректировано и далее уже будем работать со значением NaN.

В столбце rating есть два значения 'K-A' и 'E10+' - это два значения носящие один и тот же смысл (первоначально "K-A" ("Kids to Adults"), потом было переименовано на E10+), поэтому целесообазно переименовать 'K-A' в 'E10+' для удобного и более точного анализа в дальнейшем.

```python
# Функция для замены значения 'K-A' на 'E10+'
def replace_wrong_ratings(wrong_ratings, correct_rating):
    data.rating = data.rating.replace(wrong_ratings, correct_rating)
replace_wrong_ratings('K-A', 'E10+')
data.rating.value_counts()
```

Внесем небольшие изменения в наименования значений столбца platform для того, чтобы было большее понимание о каких именно игровых платформах идет речь.

```python
# Функция для замены значения аббревиатур на полное название для понимания, что это за игровая платформа.
def replace_wrong_platforms(wrong_platforms, correct_platform):
    data.platform = data.platform.replace(wrong_platforms, correct_platform)
replace_wrong_platforms('2600', 'Atari 2600')
replace_wrong_platforms('Wii', 'Nintendo Wii')
replace_wrong_platforms('NES', 'Nintendo NES')
replace_wrong_platforms('GB', 'GB-60')
replace_wrong_platforms('DS', 'Nintendo DS')
replace_wrong_platforms('X360', 'Xbox 360')
replace_wrong_platforms('SNES', 'Nintendo SNES')
replace_wrong_platforms('GBA', 'Game Boy Advance')
replace_wrong_platforms('3DS', 'Nintendo 3DS')
replace_wrong_platforms('N64', 'Nintendo 64')
replace_wrong_platforms('XB', 'Xbox')
replace_wrong_platforms('XOne', 'Xbox One')
replace_wrong_platforms('WiiU', 'Nintendo Wii U')
replace_wrong_platforms('GC', 'Nintendo GameCube')
replace_wrong_platforms('GEN', 'SEGA Retro Genesis')
replace_wrong_platforms('DC', 'Dendy Classic')
replace_wrong_platforms('PSV', 'PS Vita')
replace_wrong_platforms('SAT', 'Sega Saturn')
replace_wrong_platforms('WS', 'WonderSwan')
replace_wrong_platforms('TG16', 'NEC TurboGrafx')
replace_wrong_platforms('3DO', 'Panasonic 3DO')
replace_wrong_platforms('GG', 'Game Gear')
replace_wrong_platforms('PCFX', 'NEC PC-FX')
#проверяем уникальные значения в столбце
display(data['platform'].unique())
```

Посчитаем сколько игр вышло в разные года по разным платформам. Для этого сделаем 2 выгрузки: - первая выгрузка покажет нам сумму игр в группировке по игровой платформе и годам. - вторая выгрузка - общее количество игр по годам.

```python
years_df = (
    data.groupby("year_of_release")[["sum_sales_game"]]
    .sum()
    .join(data.groupby("year_of_release")[["name"]].count())
)
years_df.columns = ["global_sales", "number_of_games"]
years_df.head(5)
```

####  ТОП-5 платформ 2015-2016 с суммарными значениями продаж по регионам и определяем доли продаж

```python
#выведем ранее сохраненный датасет с самыми популярными платформами на 2015,2016 гг
# суммируем столбцы продаж и добавим результат в столбец sum_sales_game
list =['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'sum_sales_game']
data7 = data2.groupby('platform')[list].sum().sort_values('sum_sales_game', ascending=False).reset_index(drop=False).head(5)
data7
```

Добавим столбцы с расчитанными долями продаж по каждому региону

```python
# приводим проценты долей к целочисленному типу
data7['na_sales_share'] = round(data7.na_sales / data7.sum_sales_game * 100, 0).astype(int)
data7['eu_sales_share'] = round(data7.eu_sales / data7.sum_sales_game * 100, 0).astype(int)
data7['jp_sales_share'] = round(data7.jp_sales / data7.sum_sales_game * 100, 0).astype(int)
data7['other_sales_share'] = round(data7.other_sales / data7.sum_sales_game * 100, 0).astype(int)
data7['sum_sales_share'] = round(data7.sum_sales_game / data7.sum_sales_game * 100, 0).astype(int)
data7
```

Сделаем графическую визуализацию и посмотрим, доли популярности в регионах наиболее удачных по миру платформ:

```python
keys = ['Северная Америка','Европа','Япония','Другие страны']
palette_color = sns.color_palette('bright')
colors = sns.color_palette('pastel')[0:5]
for index, row in data7.iterrows():
    bar_numbers = [row.na_sales_share,row.eu_sales_share,row.jp_sales_share,row.other_sales_share]
    plt.title('Доли продаж по платформе ' + row.platform)
    plt.pie(bar_numbers, labels=keys, colors=palette_color, autopct='%.0f%%')
    plt.show()
```

Мы можем сделать вывод о том, что для удачного целевого маркетинга и рекламы,и как следует, последующего увеличения продаж,  необходимо не только принять во внимание общемировые значения, но сделать акценты в каждом конретном регионе именно на игры на тех платформах, которые максимально популярны в конкретном регионе.

## Проверка гипотез

T-критерий Стюдента (T-Test, T-критерий) — это статистический тест, который используется для сравнения Средних арифметических (Mean) двух групп. Он часто используется при проверке Нулевой гипотезы (Null Hypothesis).

T-Test можно использовать только при сравнении средних значений двух групп (так называемое попарное сравнение). 

Тест предполагает, что данные:

независимы
распределены (приблизительно) нормально
имеют одинаковую величину дисперсии в каждой сравниваемой группе (т. н. однородность дисперсии)
Если ваши данные не соответствуют этим предположениям, вы можете попробовать непараметрическую альтернативу t-критерию, например, Тест знаковых рангов Уилкоксона (Wilcoxon Signed-Rank Test) для данных с неравными Дисперсиями (Variance).

Какой тип T-теста следует использовать?
При выборе типа T-теста вам необходимо учитывать две вещи: относятся ли сравниваемые группы к одной Генеральной совокупности (Population) или к двум разным.

Одновыборочный, двухвыборочный или парный T-критерий? Если Выборки (Sample) берут начало из одной генеральной совокупности (например, измерение до и после экспериментального лечения), выполняют Парный T-тест (Paired T-Test).

Если группы происходят из двух разных совокупностей (например, людей из двух разных городов), выполняют Независимый T-тест (Independent T-Test).

Если есть одна группа, сравниваемая со стандартным значением , выполняют Одновыборочный T-тест (One-Sample T-Test).

Проверим две гипотезы:

- cредние пользовательские рейтинги платформ Xbox One и PC одинаковые

- cредние пользовательские рейтинги жанров Action и Sports разные

Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Для проверки гипотезы "средние пользовательские рейтинги платформ Xbox One и PC одинаковые" в качестве нулевой и альтернативной гипотезы мы взяли следующее:

###### Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Для проверки гипотезы "средние пользовательские рейтинги платформ Xbox One и PC одинаковые" в качестве нулевой и альтернативной гипотезы мы взяли следующее:
H0: средние рейтинги по платформам одинаковые
H1: средние рейтинги по платформам разные

```python
#выводим две выборки по необходимым платформам
data_xbox = data2[data2.platform == 'Xbox One']
data_pc = data2[data2.platform == 'PC']
# создаem две выборки конкретных столбцов с пользовательскими рейтингами для указанных платформ.
xbox_1 = data_xbox['user_score']
pc_1 = data_pc['user_score']
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше alpha - отвергнем гипотезу
results = st.ttest_ind(xbox_1,pc_1)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Следовательно, средние пользовательские рейтинги платформ Xbox One и PC одинаковые, и таким образом изменение объема предложения по данным платформам не будет иметь существенного влияния на рейтинги и последующие продажи.  

###### Гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

H1: средние рейтинги по жанрам разные
