#### Решение с использованием алгебраических вычислений

```python
# создание функции умножения матрицы и вывода результата
def multiply_matrix (A, B):
    res = [[sum(a * b for a, b in zip(A_row, B_col))
        for B_col in zip(*B)] for A_row in A]
    #print ("Результат умножения матриц A и B : \n  ")
    for r in res:
         print(r)
```

##### решение 1 примера

```python
A = [[6, -8, 3]]
B = [[1],
     [1],
     [4]]
print ("Ответ по результатам умножения матриц в примере №1: \n ")
multiply_matrix (A, B)
```

##### решение 2 примера

```python
A = [[-3, -7],
     [-10, 6],
     [0, -6],
     [-1, -3]]
B = [[-8, 9],
    [4, -2]]
print ("Ответ по результатам умножения матриц в примере №2: \n ")
multiply_matrix (A, B)
```

#### Решение средствами интерфейса для умножения матриц

```python
# Создание вспомогательных функция для умножения матриц

MAX = 100
def matrixPrint(M, row, col):
    for i in range(row):
        for j in range(col):
            print(M[i][j], end=" ")
        print()

# функция проверки на возможность умножения и перемножения матриц 
def matrixMultiply(row1, col1, m1, row2, col2, m2):
    res = [[0 for i in range(MAX)] for j in range(MAX)]
    if(row2 != col1):
        print("Умножение матриц невозможно из-за несоблюдения размерности")
        return
    for i in range(row1):
        for j in range(col2):
            res[i][j] = 0
            for k in range(row2):
                res[i][j] += m1[i][k] * m2[k][j]
    print("Результат умножения матриц:")
    matrixPrint(res, row1, col2)
```

```python
# функция перемножения матриц (интерфейс)
def matrix_multiply2():
    # main
    if __name__ == "__main__":
        m1 = [[0 for i in range(MAX)] for j in range(MAX)]
        m2 = [[0 for i in range(MAX)] for j in range(MAX)]
        row1 = int(input("Введите количество строк матрицы 1: "))
        col1 = int(input("Введите количество рядов матрицы 1: "))
        print("Ввведите значения элементов матрицы 1: ")
        for i in range(row1):
            for j in range(col1):
                m1[i][j] = int(input("m1[" + str(i) + "][" + str(j) + "]: "))

        row2 = int(input("Введите количество строк матрицы 2: "))
        col2 = int(input("Введите количество рядов матрицы 2: "))

        print("Ввведите значения элементов матрицы 2: ")
        for i in range(row2):
            for j in range(col2):
                m2[i][j] = int(input("m2[" + str(i) + "][" + str(j) + "]: "))
        print("Матрица 1: ")
        matrixPrint(m1, row1, col1)
        print("Матрица 2: ")
        matrixPrint(m2, row2, col2)
        matrixMultiply(row1, col1, m1, row2, col2, m2)
```
