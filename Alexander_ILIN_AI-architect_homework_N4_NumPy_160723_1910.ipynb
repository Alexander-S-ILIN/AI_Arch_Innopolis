## Задача №1



### Решение

Создание массивов NumPy для матриц A, B и С:

```python
matrix_A = np.array([[1, 2, 1], 
                     [-2, 1, 1]])
```

```python
matrix_B = np.array([[3, 1], 
                     [-1, 0],
                     [2, 4]])
```

```python
matrix_C = np.array([[-2, 2], 
                     [1, -1],
                     [1, 2]])
```

```python
print ("Матрица А:\n", matrix_A)
print ("Матрица B:\n", matrix_B)
print ("Матрица C:\n", matrix_C)
```

#### Теоретическое обоснование наличия возможности сложения матриц
NumPy в Питоне позволяет складывать только матрицы одинаковых размеров.
Матрицы складываются с помощью сложения всех элементов массива с одинаковыми индексами. Матрица с суммами этих элементов является результатом сложения.

#### Создание функции вычисления примеров

```python
# на вход подаются 2 матрицы, вторая из которых сначала умножается на скаляр и потом производится попытка сложения матриц
def matrix_sum(matrix_1, matrix_2, scalyar):
    try:
        if matrix_1.shape == matrix_2.shape:
            print (matrix_1 + (scalyar * matrix_2)) 
        else:
            print ('''Невозможно осуществить операцию сложения:
            сложение матриц возможно исключительно при равенстве их размерности.''')
            
    except:
        print ('''Невозможно осуществить операцию. 
                    Проверьте входные параметры. На вход функции сложения матриц должны быть поданы 2 матрицы''')
            
    return
```

```python
print('Результат решения примера:  A + 2B: \n') 
matrix_sum(matrix_A, matrix_B, 2)
```

```python
print('Результат решения примера:  B + 2C: \n') 
matrix_sum(matrix_B, matrix_C, 2)
```

```python
print('Проверка отработки неверно поданных входящих параметров на функцию вычисления примеров: \n') 
matrix_sum('dgsgadgag', matrix_B, 2)
matrix_sum(False, matrix_B, 2)
matrix_sum(55657, matrix_B, 2)
```

## Задача №3


### Решение

#### Формирование векторов

```python
v1 = np.linspace(0, 1.4, 15)
v1
```

```python
v2 = np.linspace(1,6,6)
v2
```

#### Преобразование векторов в матрицы , заданных размеров

```python
matrix1 = v1.reshape(5,3)
print('По результатам преобразования получаем матрицу 5 х 3: \n', matrix1)
```

```python
matrix2 = v2.reshape(3,2)
print('По результатам преобразования получаем матрицу 3 х 2: \n', matrix2)
```

#### Перемножение матриц

#####  1 вариант : через функцию np.dot

```python
print('Результат перемножения матриц через функцию np.dot : \n', np.dot(matrix1, matrix2))
```

#####  2 вариант : через оператор @

```python
print('Результат перемножения матриц через оператор @ : \n', matrix1@matrix2)
```
